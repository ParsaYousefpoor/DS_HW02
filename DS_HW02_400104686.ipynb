```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("manchunhui/world-development-indicators")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import os

# Assuming 'path' variable holds the directory where the dataset is downloaded
# Replace with your actual path if it's different
# Example: path = "/content/world-development-indicators"

# Check if the path exists
if 'path' in locals() and os.path.exists(path):
  # Find CSV files within the directory
  csv_files = [f for f in os.listdir(path) if f.endswith('.csv')]

  if csv_files:
      # Load the first CSV file found (modify as needed if there are multiple files)
      file_path = os.path.join(path, csv_files[0])
      try:
          df = pd.read_csv(file_path)
          print("Dataset loaded successfully.")
          #print(df.head())  # Display first few rows of the dataframe
      except Exception as e:
          print(f"Error loading CSV file: {e}")
  else:
      print("No CSV files found in the specified directory.")
else:
  print("The provided path does not exist. Please check the path or run the kaggle dataset download code first.")
```
